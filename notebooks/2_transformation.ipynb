# Salary Parsing


```python
PHP_TO_USD = 1 / 57.75 # as of feb 26, 2026
```

## Currency Detection
Detects whether a raw salary string is denominated in USD or PHP based on symbols and keywords.

```python
def detect_currency(raw: str) -> str:
    s = raw.replace("?", "₱")

    has_php = bool(re.search(r"PHP|php|Php|Peso|peso|Philippine|₱", s))
    has_usd = bool(re.search(r"USD|usd|Usd|US\$|U\.S\.\$|\$", s))

    if has_usd:
        return "USD"
    if has_php:
        return "PHP"
    return "USD"
```

## Number Extraction Helpers
Utility functions to normalize raw strings and extract numeric values — either from the full string or after a specific currency symbol.

```python
def _normalise_raw(raw: str) -> str:
    s = raw.replace("?", "₱")
    s = re.sub(r"(\d),(\d{3})", r"\1\2", s)  # 40,000 → 40000
    return s


def _extract_numbers(s: str) -> list:
    return [float(n) for n in re.findall(r"\b\d+(?:\.\d+)?\b", s)]


def _extract_by_symbol(norm: str, symbol_pat: str) -> list:
    return [
        float(m.group(1))
        for m in re.finditer(symbol_pat + r"\s*(\d+(?:\.\d+)?)", norm, re.IGNORECASE)
    ]
```

## Salary Type Detection
Detects salary frequency (hourly, daily, weekly, monthly, annually) from explicit keywords in the raw string, with fallback heuristics based on the numeric value and currency.

```python
_EXPLICIT_TYPE_PATTERNS = [
    (r"\b(?:per\s*hour|/\s*hour|/\s*hr\b|per\s*hr\b|p/?h\b|hourly)\b", "hourly"),
    (r"\b(?:per\s*day|/\s*day|daily)\b", "daily"),
    (r"\b(?:per\s*week|/\s*week|/\s*wk\b|weekly)\b", "weekly"),
    (r"\b(?:per\s*year|/\s*year|annually|annual|per\s*annum|/\s*yr\b)\b", "annually"),
    (
        r"\b(?:per\s*month|/\s*month|/\s*mo\b|monthly|per\s*mo\b|p/?m\b|pcm\b)\b",
        "monthly",
    ),
    (r"/\s*mo\b", "monthly"),
    (r"/\s*hr\b", "hourly"),
    (r"/\s*wk\b", "weekly"),
    (r"/\s*yr\b", "annually"),
    (r"\d\s+hours?\b", "hourly"),
    (r"\b(?:hr|hrs)\b", "hourly"),
    (r"\bmo\b", "monthly"),
    (r"\bwk\b", "weekly"),
]

_USD_HEURISTICS = [
    (0, 20, "hourly"),
    (20, 300, "weekly"),
    (300, 7_000, "monthly"),
    (7_000, 1e9, "annually"),
]

_PHP_HEURISTICS = [
    (0, 1_000, "daily"),
    (1_000, 10_000, "weekly"),
    (10_000, 500_000, "monthly"),
    (500_000, 1e9, "annually"),
]


def detect_salary_type(raw: str, avg_usd, currency: str) -> str:
    text = raw.lower()
    for pattern, stype in _EXPLICIT_TYPE_PATTERNS:
        if re.search(pattern, text, re.IGNORECASE):
            return stype
    if avg_usd is None:
        return "unknown"
    table = _USD_HEURISTICS if currency == "USD" else _PHP_HEURISTICS
    ref = avg_usd if currency == "USD" else avg_usd / PHP_TO_USD
    for lo, hi, stype in table:
        if lo <= ref < hi:
            return stype
    return "monthly"
```

## Core Salary Parser
Parses a single raw salary string into a structured dict with min, max, average, USD-equivalent, currency, and salary type. Handles mixed-currency strings, filters noise numbers, and collapses outliers.

```python
def parse_salary(raw) -> dict:
    empty = {
        "min_salary": None,
        "max_salary": None,
        "average_salary": None,
        "usd_salary": None,
        "currency": None,
        "salary_type": None,
    }

    if raw is None or (isinstance(raw, float) and np.isnan(raw)):
        return empty
    raw = str(raw).strip()
    if not raw:
        return empty

    norm = _normalise_raw(raw)
    currency = detect_currency(raw)

    has_php = bool(re.search(r"PHP|php|Php|Peso|peso|₱|\?", raw))
    has_usd = bool(re.search(r"USD|usd|Usd|US\$|\$", raw))

    nums: list

    if has_php and has_usd:
        usd_nums = _extract_by_symbol(norm, r"\$")
        php_nums = _extract_by_symbol(norm, r"(?:PHP|Php|php|₱)")
        if usd_nums:
            nums = usd_nums
            currency = "USD"
        elif php_nums:
            nums = php_nums
            currency = "PHP"
        else:
            nums = _extract_numbers(norm)
    else:
        nums = _extract_numbers(norm)

    if currency == "USD":
        large = [n for n in nums if n >= 50]
        nums = large if large else [n for n in nums if n >= 0.25]
    else:
        large = [n for n in nums if n >= 500]
        nums = large if large else [n for n in nums if n >= 50]

    if len(nums) > 2:
        nums_s = sorted(nums)
        best: list = nums_s[:2]
        for i in range(len(nums_s)):
            for j in range(i + 2, len(nums_s) + 1):
                w = nums_s[i:j]
                if w[-1] / (w[0] + 1e-9) < 25 and len(w) >= len(best):
                    best = w
        nums = best

    if not nums:
        return {**empty, "currency": currency, "salary_type": "unknown"}

    min_sal = min(nums)
    max_sal = max(nums)
    avg_sal = (min_sal + max_sal) / 2
    usd_avg = avg_sal * PHP_TO_USD if currency == "PHP" else avg_sal

    salary_type = detect_salary_type(raw, usd_avg, currency)

    return {
        "min_salary": round(min_sal, 2),
        "max_salary": round(max_sal, 2),
        "average_salary": round(avg_sal, 2),
        "usd_salary": round(usd_avg, 2),
        "currency": currency,
        "salary_type": salary_type,
    }
```

## DataFrame Transformer
Applies `parse_salary` across an entire DataFrame column and appends the six parsed fields as new columns.

```python
def transform_salary_column(
    df: pd.DataFrame, salary_col: str = "salary"
) -> pd.DataFrame:
    parsed_rows = df[salary_col].apply(parse_salary)
    parsed_df = pd.DataFrame(parsed_rows.tolist(), index=df.index)
    return pd.concat([df, parsed_df], axis=1)
```

```python
SKILL_MAP_JSON = {
    "Technology & Development": {
        "Web Development": [
            "React JS",
            "Vue JS",
            "Vue 2",
            "Vue 3",
            "Angular JS",
            "Next JS",
            "Nuxt JS",
            "Svelte",
            "Node JS",
            "Express JS",
            "Laravel",
            "Django",
            "Flask",
            "FastAPI",
            "PHP",
            "CodeIgniter",
            "CakePHP",
            "Ruby on Rails",
            "ASP.NET",
            ".Net",
            "Spring Boot",
            "Nest JS",
            "Backbone JS",
            "Alpine.js",
            "HTML",
            "CSS",
            "SCSS",
            "SASS",
            "TailwindCSS",
            "Bootstrap",
            "Semantic UI",
            "jQuery",
            "Javascript",
            "Typescript",
            "Livewire",
            "Vite",
            "Webflow",
            "Wordpress",
            "Elementor",
            "Wix",
            "Squarespace",
            "Woocommerce",
            "Magento",
            "PrestaShop",
            "OpenCart",
            "Duda",
            "Carrd",
            "Gutenberg",
            "Godaddy Website Builder",
            "Odoo Website Builder",
            "PSD to HTML",
            "Web Programming",
            "Web Development",
            "Fullstack",
            "Web Scraping",
            "Data scraping",
            "CMS",
            "Headless CMS",
            "Netlify CMS",
            "ExpressionEngine",
            "Umbraco",
            "Drupal",
            "Solodev",
            "OutSystems",
            "Bubble.io",
            "APPSHEET",
            "CSS",
            "Material Components Web",
            "MVP.css",
            "Open Props",
            "Vanilla Framework",
            "LESS",
            "Liquid",
            "REST API",
            "GraphQL",
            "JSON",
            "Cpanel",
            "FTP and SFTP",
            "DNS Management",
            "Nginx",
            "Website Builder",
            "Google Sites",
            "Contentstack",
            "CoreMedia Content Cloud",
            "FlutterFlow",
        ],
        "Mobile Development": [
            "React Native",
            "Flutter",
            "Swift",
            "Swiftui",
            "Kotlin",
            "Ionic",
            "iOS Development",
            "Android Development",
            "Objective-C",
            "Dart",
            "Mobile Development",
            "Mobile Service",
            "Mobile Marketing",
            "Maui",
        ],
        "Software Engineering": [
            "Python",
            "Java",
            "C#",
            "C++",
            "Ruby",
            "GoLang",
            "R",
            "PHP",
            "Bash",
            "Powershell",
            "Visual Basic",
            "VB.NET",
            "SQL",
            "Software Development Project Manager",
            "Software",
            "Scrum",
            "Agile Methodology",
            "Git",
            "GitHub",
            "Atlassian",
            "Jira",
            "Asana",
            "Management and Developer Tools",
            "Dev Tools - Pipelines",
            "Dev Tools - Visual Studio",
            "Dev Tools - CodeDeploy",
            "Dev Tools - CodeCommit",
            "Dev Tools - CodeGuru",
            "Dev Tools - Cloud Development Kit",
            "Dev Tools - Microsoft Dev Box",
            "Apps Script",
            "PowerApps",
            "Power Automate",
            "ServiceNow",
            "Uipath",
            "Robotic Process Automation",
        ],
        "DevOps & Cloud": [
            "AWS",
            "AWS Lambda",
            "Azure",
            "Google Cloud",
            "Cloud Computing",
            "Cloud Services",
            "Cloud Engineer",
            "Devops",
            "CICD",
            "Docker",
            "Kubernetes",
            "Terraform",
            "Ansible",
            "Linux Administrator",
            "Windows Administrator",
            "Server",
            "Systems Administration",
            "Network Administration",
            "Network Monitoring",
            "Network Security",
            "Network Troubleshooting",
            "Networking",
            "Networking - Traffic Manager",
            "Networking - Azure Private Link",
            "Compute - VMWare Solution",
            "Compute - EC2 Auto Scaling",
            "Internet of Things - IoT Core",
            "Storage and Databases",
            "Google Workspace",
            "Microsoft 365",
            "SharePoint",
            "Active Directory",
            "Security - Azure Active Directory",
            "Security - Microsoft Defender for Cloud",
            "Security - Microsoft Sentinel",
            "Identity & Security",
            "Cybersecurity",
            "VPN",
            "VPN - OpenVPN",
            "Firewalls",
            "Threat Hunting",
            "Application Performance Management",
            "Snowflake",
            "Mail - Postfix",
            "Mail - Sendmail",
        ],
        "Database & Data Engineering": [
            "MySQL",
            "PostgreSQL",
            "SQL Server",
            "MongoDB",
            "MariaDB",
            "Oracle",
            "Redis",
            "Elasticsearch",
            "Supabase",
            "Bigquery",
            "Firebase",
            "Database Management",
            "Databases",
            "Database - DynamoDB",
            "Database - SQL Edge",
            "Mssql",
            "NetSuite",
            "Data Engineering",
            "EDI",
            "Data Management",
            "Data Collection",
            "Data Gathering",
            "Data Capture",
            "Admin Data Processing",
            "Web3.js",
            "Blockchain",
            "Ethereum",
        ],
        "AI & Automation": [
            "AI Tools",
            "AI Automation",
            "Chatbot",
            "ChatGPT",
            "Artificial Intelligence",
            "Machine Learning",
            "Prompt Engineering",
            "Midjourney",
            "Stable Diffusion",
            "Make.com",
            "Zapier",
            "Marketing Automation",
            "Workflow",
            "Robotic Process Automation",
            "Automation",
            "Integration",
            "API Integration",
            "API Development",
            "Stripe Api",
            "Paypal Api",
            "REST API",
            "API Testing",
            "Ethers.js",
        ],
        "QA & Testing": [
            "Quality Assurance",
            "Software QA Testing",
            "Manual Testing",
            "Functional Testing",
            "Regression Testing",
            "Test Automation",
            "Usability Testing",
            "API Testing",
            "Bug Tracking",
            "AB Testing",
            "Playwright",
            "Cypress",
            "Selenium",
            "Quality Control",
            "Error Detection",
            "System Analysis",
        ],
    },
    "Design & Creative": {
        "Graphic Design": [
            "Graphic Design",
            "Canva",
            "Photoshop",
            "Illustrator",
            "Indesign",
            "Coreldraw",
            "Figma",
            "Adobe XD",
            "Procreate",
            "Clip Studio Paint",
            "Print Design",
            "Infographics",
            "Logo Design",
            "T-Shirt Design",
            "Packaging & Brand Idenity Design",
            "Advertising And Marketing Design",
            "Design",
            "Visual Design",
            "Visual Creation",
            "Cartooning",
            "Illustration",
            "Drawing",
            "Painting",
            "Art & Comic Design",
            "Font & Typographic Design",
            "ZBrush",
            "Annotation",
        ],
        "UI/UX Design": [
            "UI UX Design",
            "User Experience Design",
            "User Interface",
            "User Experience Research",
            "Wireframing",
            "Prototyping",
            "Interaction Design",
            "Design Thinking",
            "Design Strategy",
            "Information Architecture",
            "Figma",
            "Adobe XD",
            "Design Ops",
            "Usability Testing",
        ],
        "Video Production & Editing": [
            "Video Editing",
            "Video Content Creation",
            "Video Marketing",
            "Social Media Video Editing",
            "Adobe Premiere Pro",
            "Davinci Resolve",
            "Apple Final Cut Pro",
            "CapCut",
            "Clipchamp",
            "Movavi Video Editor Plus",
            "Corel Videostudio Ultimate",
            "Wondershare Filmora",
            "Adobe Premiere Elements",
            "Film Editing",
            "Movie Editing",
            "Color Grading",
            "Captioning",
            "Encoding",
            "Videography",
            "Descript",
        ],
        "Photo Editing & Photography": [
            "Photo Editing",
            "Photography",
            "Photo Retouching",
            "Lightroom",
            "Photoshop",
            "Capture One",
            "Photo Stitching",
        ],
        "Motion Graphics & Animation": [
            "Motion Graphics",
            "After Effects",
            "Animation",
            "2d Animation",
            "3D animation",
            "Storyboarding",
            "3D blender",
            "3DS Max",
            "Autodesk Maya",
            "3D unity",
            "Unity",
            "Unreal Engine",
            "Vray",
            "Lumion",
            "Twinmotion",
            "D5 Render",
            "Enscape",
        ],
        "3D Design & Modeling": [
            "3D Design",
            "3D modeling",
            "3D illustration",
            "3D presentation",
            "3D autocad",
            "Sketchup",
            "Autodesk Fusion 360",
            "Autodesk Inventor",
            "Solidworks",
            "Onshape",
            "Parametric modeling",
            "Revit",
            "Autocad",
            "Civil 3D",
            "Auto Cad",
            "Microstation",
            "ZBrush",
            "Cabinet Vision",
            "Navisworks",
        ],
        "Branding & Identity": [
            "Branding",
            "Brand Management",
            "Brand Analysis",
            "Packaging & Brand Idenity Design",
            "Logo Design",
            "Creative Direction",
            "Creative Advertising",
        ],
        "Music & Audio Production": [
            "Music Production",
            "Sound Design",
            "Audio Editing",
            "Voice Over",
            "Voice Acting",
            "Podcast Content Creation",
            "Songwriting",
            "Flm Scoring",
            "FL Studio",
            "Ableton Live",
            "Garageband",
            "Audacity",
            "Audiotool",
            "Adobe Audition Cc",
            "Music Production",
        ],
    },
    "Marketing & Advertising": {
        "Digital Marketing": [
            "Digital Marketing",
            "Marketing",
            "Marketing Analytics",
            "Marketing Management",
            "Performance Marketing",
            "Marketing Automation",
            "Search Engine Marketing",
            "Advertising",
            "Ads Management",
            "Ads Creation",
            "Creative Advertising",
            "Scientific Advertising",
            "Advertising Management",
        ],
        "Social Media Marketing": [
            "Social Media Marketing",
            "Social Media Management",
            "Social Media Content Creation",
            "Social Media Engagement",
            "Social Media Moderation",
            "Facebook marketing",
            "Facebook Ads",
            "Instagram marketing",
            "LinkedIn Marketing",
            "Youtube Marketing",
            "Youtube Ads",
            "Tiktok Ads",
            "Twitter Ads",
            "Pinterest Ads",
            "Snapchat Ads",
            "Meta Business Suite",
            "Meta Pixel",
            "Post Scheduling",
            "Hootsuite",
            "Influencer Marketing",
            "Influencer Content Creation",
            "Community Moderation",
            "Forum Community Moderation",
            "Online Forum Posting",
        ],
        "SEO & SEM": [
            "SEO",
            "Local SEO",
            "SEO Analysis",
            "On-Page",
            "Off-Page",
            "Keyword Research",
            "Keyword Management",
            "Backlinks Management",
            "Link Building",
            "Ahrefs",
            "Semrush",
            "Google Search Console",
            "Google Analytics",
            "GA4",
            "Google Tag Manager",
            "Google Data Studio",
            "Looker Studio",
            "Ranking Management",
            "Private Blog Network",
            "Search Engine Marketing",
        ],
        "Paid Advertising": [
            "Google Ads",
            "Facebook Ads",
            "Amazon Ads",
            "Amazon PPC",
            "Tiktok Ads",
            "Twitter Ads",
            "Pinterest Ads",
            "Snapchat Ads",
            "Bing Ads",
            "Microsoft Ads",
            "Yahoo Gemini",
            "PPC",
            "Media Buying",
            "Campaign Optimization",
            "Campaign Building",
            "Performance Tracking",
            "Funnelish",
            "Funnel Design",
            "Clickfunnels",
            "Kartra",
        ],
        "Email Marketing": [
            "Email Marketing",
            "eMail Marketing",
            "Email Copywriting",
            "Email Handling",
            "Email Management",
            "Email Support",
            "Klaviyo",
            "Mailchimp",
            "MailerLite",
            "Active Campaign",
            "Convertkit",
            "Flodesk",
            "Benchmark Email",
            "Omnisend",
            "EmailOctopus",
            "Ontraport",
            "Keap",
        ],
        "Content Marketing": [
            "Content Creation",
            "Content Management",
            "Content Planning",
            "Social Media Content Creation",
            "Blogging",
            "Microblogging",
            "Microblog Content Writing",
            "Podcast Content Creation",
            "Thumbnail & Social Media Design",
            "Trend Monitoring",
            "Audience Analysis",
            "Audience Engagement",
        ],
        "Influencer & Affiliate Marketing": [
            "Influencer Marketing",
            "Influencer Content Creation",
            "Affiliate Marketing",
        ],
        "Analytics & Performance": [
            "Marketing Analytics",
            "Google Analytics",
            "GA4",
            "Google Tag Manager",
            "Google Data Studio",
            "Looker Studio",
            "Power BI",
            "Tableau",
            "Qlik Sense",
            "Analytics",
            "Data analysis",
            "Data Analytics",
            "Data Visualization",
            "Performance Tracking",
            "AB Testing",
            "Market Research",
            "Competitor Analysis",
            "Brand Analysis",
            "Feed Analysis",
            "Audience Analysis",
        ],
    },
    "Sales & Business Development": {
        "Sales": [
            "Sales",
            "Inbound Sales",
            "Outbound Sales",
            "Sales Management",
            "Sales Support",
            "B2B",
            "B2C",
            "Cold Calling",
            "Warm Calling",
            "Outbound Calls",
            "Telemarketing",
            "Account Acquisition",
            "Prospecting",
            "Negotiation",
            "Negotation",
        ],
        "Lead Generation & CRM": [
            "Lead Generation",
            "CRM",
            "HubSpot CRM",
            "Salesforce",
            "Zoho CRM",
            "GoHighLevel",
            "Pipedrive",
            "PipelineDeals",
            "NetSuite CRM",
            "Freshworks CRM",
            "Insightly",
            "Sugar CRM",
            "Less Annoying CRM",
            "Apptivo",
            "Bullhorn",
            "Creatio",
            "Podio",
            "Apollo",
            "Client Relationship Management",
            "Customer Relationship Management",
            "Account Management",
            "Account Retention",
        ],
        "Business Development": [
            "Business development",
            "Business Analysis",
            "Business Management",
            "Strategic Planning",
            "Networking",
            "Acquisition",
            "Acquisition Management",
            "Skip Tracing",
        ],
    },
    "Writing & Content": {
        "Copywriting & Editing": [
            "Copywriting",
            "Email Copywriting",
            "Website Copywriting",
            "Copyediting",
            "Proofreading",
            "Editing",
        ],
        "Content Writing": [
            "Content Creation",
            "Article Writing",
            "Blog",
            "Blogging",
            "Microblog Content Writing",
            "News Writing",
            "Journalism",
            "Listicles",
            "Case Studies Writing",
            "Webinar Writing",
            "Online Forum Posting",
            "Ghost Writing",
            "eBook Writing",
            "Kindle Direct Publishing (KDP)",
        ],
        "Technical & Specialized Writing": [
            "Technical Writing",
            "Research Writing",
            "Academic Writing",
            "Scientific Research",
            "Scriptwriting",
            "Storytelling",
            "Legal Writing",
            "Policy Writing",
            "Sop Writing",
            "Documentation",
            "Patent Drafting",
            "Medical Scribing",
        ],
        "Creative Writing": [
            "Creative Writing",
            "Scriptwriting",
            "Storytelling",
            "Ghost Writing",
            "eBook Writing",
            "Songwriting",
        ],
    },
    "Administrative & Operations": {
        "Virtual Assistance & Admin": [
            "Virtual Assistant",
            "Personal Assistant",
            "Administrative Support",
            "Administrative Management",
            "Executive Assistance",
            "Office and Administration",
            "Back Office Support",
            "Calendar Management",
            "Email Management",
            "Email Handling",
            "Scheduling",
            "Meeting Management",
            "Appointment Setting",
            "Travel Planning",
            "Travel Management",
            "File Management",
            "Records Management",
            "Data Entry",
            "Typing",
            "Transcription",
            "Note Taking",
            "Template Creation",
            "Google Docs",
            "Google Sheets",
            "Google Drive",
            "Google Slides",
            "Google Forms",
            "Microsoft Excel",
            "Microsoft Word",
            "Microsoft Power Point",
            "Microsoft Outlook",
            "Google Suite",
            "Google Workspace",
        ],
        "Project & Operations Management": [
            "Project Management",
            "Operation Management",
            "Operations",
            "Process Improvement",
            "Process Mapping",
            "Workflow",
            "Coordination",
            "Logistics Management",
            "Logistics Coordination",
            "Freight Forwarding",
            "Supply Chain Management",
            "Supply Management",
            "Procurement",
            "Purchasing",
            "Inventory Management",
            "Order Management",
            "Order processing",
            "Order Tracking",
            "Order Fulfillment",
            "Dispatching",
            "Capacity Planning",
            "Strategic Planning",
            "Budget Management",
            "Costing",
            "Estimate",
            "Forecasting",
        ],
        "Customer Support": [
            "Customer Support",
            "Customer Service",
            "Customer Support Management",
            "Chat Support",
            "Phone Support",
            "Email Support",
            "Technical Support",
            "Helpdesk Support",
            "Zendesk",
            "Gorgias",
            "Freshdesk",
            "HelpScout",
            "Intercom",
            "Onboarding",
            "Community Moderation",
        ],
        "Tools & Platforms": [
            "Clickup",
            "Asana",
            "Trello",
            "Monday",
            "Notion",
            "Airtable",
            "Smartsheet",
            "Basecamp",
            "Jira",
            "Scoro",
            "Podio",
            "WORKetc",
            "Accelo",
            "Teamwork",
            "Jotform",
        ],
    },
    "Finance & Accounting": {
        "Bookkeeping & Accounting": [
            "Bookkeeping",
            "Accounting",
            "Online Bookkeeping",
            "Accounting Management",
            "Financial Accounting",
            "Cost Accounting",
            "Investment Accounting",
            "General Ledger",
            "Accounts Payable",
            "Accounts Receivable",
            "Accounts Reconciliations",
            "Bank Reconciliation",
            "Xero",
            "Quickbooks",
            "MYOB",
            "Netsuite Accounting Software",
            "Sage Intacct",
            "SAP",
            "SAP ERP",
            "SAP S4HANA",
            "Odoo",
            "Acumatica",
        ],
        "Financial Planning & Analysis": [
            "Financial Planning",
            "Financial Analysis",
            "Financial Reporting",
            "Financial Forcasting",
            "Forecasting",
            "Budget Management",
            "Cash Flow Management",
            "Risk Analysis",
            "Risk Management",
            "Risk Assessment",
            "Investment Research",
            "Investing",
            "Forex Trading",
            "Stock Trading",
            "Technical Analysis",
            "Underwriting",
        ],
        "Tax, Payroll & Billing": [
            "Tax Preparation",
            "Payroll",
            "Billing",
            "Invoicing",
            "Medical Billing",
            "Dental Billing",
            "Revenue Cycle Management",
            "Revenue Management",
            "Denial Management",
            "Claims Processing",
            "Chargeback Analysis",
            "Debt Collection",
            "Collection",
            "Payment Processing",
            "Loan Processing",
            "Mortgages",
            "Mortgage",
            "Credit Repair",
            "Insurance",
            "Insurance Verification",
            "Prior Authorization",
            "Car Insurance",
            "Home Insurance",
            "Medical Insurance",
            "Dental Insurance",
            "Internal Auditing",
            "Auditing",
            "External Audit",
        ],
    },
    "E-Commerce": {
        "Marketplace Management": [
            "Ecommerce",
            "E-Commerce",
            "E-Commerce Management",
            "Amazon",
            "FBA",
            "Online Arbitrage",
            "Seller Central",
            "Seller Page Management",
            "Listing Management",
            "Product Listing",
            "Catalog Management",
            "Etsy",
            "Ebay",
            "Shopify",
            "Woocommerce",
            "Bigcommerce",
            "Dropshipping",
            "Fulfillment",
            "Helium10",
            "JungleScout",
            "Keepa",
            "Seller Amp SAS",
            "Tactical Arbitrage",
            "Data Dive",
        ],
        "Product Management": [
            "Product Research",
            "Product Sourcing",
            "Wholesale Sourcing",
            "Supplier Sourcing",
            "Product Design",
            "Product Optimization",
            "Product Analysis",
            "Product Knowledge",
            "Inventory Management",
            "Order Management",
            "Order Fulfillment",
            "Order Tracking",
        ],
    },
    "Real Estate": {
        "Real Estate": [
            "Real Estate",
            "Real Estate Marketing",
            "Commercial Real Estate",
            "Gohighlevel Real Estate",
            "ISA Real Estate",
            "Disposition Real Estate",
            "REO Asset Management",
            "Transaction Coordinating",
            "Listing Coordination",
            "Property Management",
            "Short Term Rentals",
            "Airbnb",
            "Vrbo",
            "HostAway",
            "Appfolio",
            "Buildium",
            "Yardi",
            "Brivity",
            "Title Escrow",
            "Title Processing",
            "Appraisal",
            "Trustee Sales Tracking",
            "Corelogic Mls Matrix",
            "Skip Tracing",
        ]
    },
    "Human Resources & Recruiting": {
        "Recruiting": [
            "Recruiting",
            "Technical recruiting",
            "US recruiting",
            "Talent Acquisition",
            "Candidate Sourcing",
            "Candidate Screening",
            "Boolean Search",
            "Applicant Tracking Systems",
            "Staffing",
            "Labor Relations",
        ],
        "HR Management": [
            "Human Resources",
            "HR Management",
            "Employee Engagement",
            "Workforce Management",
            "Training and Development",
            "Onboarding",
            "Instructional Design",
            "Coaching",
            "Life Coaching",
            "Performance Tracking",
            "Payroll",
            "Timekeeping",
            "Credentialing",
            "Workday",
        ],
    },
    "Healthcare & Medical": {
        "Medical Administration": [
            "Medical Billing",
            "Medical Coding",
            "Medical Scribing",
            "Medical Virtual Assistant",
            "Electronic Health Record",
            "HIPAA Compliance",
            "Revenue Cycle Management",
            "Denial Management",
            "Insurance Verification",
            "Prior Authorization",
            "ICD-10",
            "Claims Processing",
            "Records Retrieval",
            "Credentialing",
        ],
        "Clinical & Health": [
            "Healthcare",
            "Patient Care",
            "Medical Services",
            "Medical Knowledge",
            "Health Education",
            "Health And Wellness",
            "Nutrition Counseling",
            "Nutrition Assessment",
            "Fitness",
            "Psychology",
            "Psychological Assessment",
            "Counseling",
            "Case Management",
        ],
    },
    "Legal & Compliance": {
        "Legal": [
            "Legal Services",
            "Legal Research",
            "Legal Writing",
            "Paralegal",
            "Case Management",
            "Patent Drafting",
            "Intellectual Property Rights",
            "Regulatory Compliance",
            "Policy Writing",
        ]
    },
    "Engineering & Architecture": {
        "Civil & Structural Engineering": [
            "Structural Engineering",
            "Civil 3D",
            "STAAD",
            "Etabs",
            "Primavera",
            "Quantity Surveying",
            "Quantity Takeoff",
            "Construction Project Manager",
            "Construction And Engineering",
            "Building Information Modeling (BIM)",
            "Bluebeam",
            "Bluebeam Revu",
            "Planswift",
            "CostX",
            "Buildxact",
            "Navisworks",
            "Tekla Structures",
        ],
        "Architectural Design": [
            "Architectural Services",
            "Revit",
            "Autocad",
            "Auto Cad",
            "Sketchup",
            "Archicad",
            "Lumion",
            "Twinmotion",
            "D5 Render",
            "Enscape",
            "2D layout",
            "3D autocad",
            "Building Information Modeling (BIM)",
        ],
        "Electrical & Other Engineering": [
            "Electrical Design",
            "Dialux",
            "Aurora Solar",
            "GIS",
            "QGIS",
            "ArcGIS",
            "Simulation",
            "Solidworks",
            "Autodesk Inventor",
            "Autodesk Fusion 360",
            "Timber Design",
            "Steel and Rigid Frames",
        ],
    },
    "Soft Skills & General": {
        "Communication": [
            "Communication",
            "English",
            "English Proficiency",
            "English Speaking",
            "Spanish Bilingual",
            "Korean Bilingual",
            "German Language",
            "Japanese Speaking",
            "Translation",
            "Public Speaking",
            "Active Listening",
            "Interpersonal",
            "Collaboration",
            "Teamwork",
            "Team Collaboration",
            "Team Leadership",
        ],
        "Professional Skills": [
            "Leadership",
            "Management",
            "Problem Solving",
            "Creative Problem Solving",
            "Critical Thinking",
            "Logical Reasoning",
            "Decision Making",
            "Analytical",
            "Conflict resolution",
            "Negotiation",
            "Emotional Intelligence",
            "Adaptability",
            "Flexibility",
            "Resourcefulness",
            "Creativity",
            "Storytelling",
            "Attention to Detail",
            "Accuracy Verification",
            "Organization",
            "Time Management",
            "Time Managment",
            "Multitasking",
            "Self-motivation",
            "Reliability",
            "Work Ethics",
            "Tech-Savvy",
            "Computer Literacy",
            "Technology Proficiency",
            "Research",
            "Analysis",
            "Planning",
            "Task Prioritization",
            "Visualization",
            "Writing",
        ],
    },
}

skill_lookup = {}
for category, subcategories in SKILL_MAP_JSON.items():
    for subcategory, skills in subcategories.items():
        for skill in skills:
            skill_lookup[skill.lower().strip()] = (category, subcategory)


def map_category_subcategory(skills_str):
    """Map a semicolon-separated skills string to a dominant (category, subcategory)."""
    if not isinstance(skills_str, str) or not skills_str.strip():
        return None, None

    skills_list = [s.strip() for s in skills_str.split(";") if s.strip()]

    mapped = []
    for skill in skills_list:
        result = skill_lookup.get(skill.lower())
        if result:
            mapped.append((skill, result[0], result[1]))

    if not mapped:
        return None, None

    from collections import Counter

    category_counts = Counter(cat for _, cat, _ in mapped)
    max_count = max(category_counts.values())

    dominant_category = None
    for _, cat, _ in mapped:
        if category_counts[cat] == max_count:
            dominant_category = cat
            break

    subcat_counts = Counter(sub for _, cat, sub in mapped if cat == dominant_category)
    max_subcat_count = max(subcat_counts.values())

    dominant_subcategory = None
    for _, cat, sub in mapped:
        if cat == dominant_category and subcat_counts[sub] == max_subcat_count:
            dominant_subcategory = sub
            break

    return dominant_category, dominant_subcategory


df[["category", "subcategory"]] = df["skills"].apply(
    lambda x: pd.Series(map_category_subcategory(x))
)

skills_idx = df.columns.get_loc("skills")
cols = df.columns.tolist()
cols.remove("category")
cols.remove("subcategory")
cols.insert(skills_idx + 1, "category")
cols.insert(skills_idx + 2, "subcategory")
df = df[cols]

df[["skills", "category", "subcategory"]].head(20)
```
